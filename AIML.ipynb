```python
from google.colab import files
uploaded = files.upload()

import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from collections import Counter

#the below few lines will be used to load and preprocess the data
iris=load_iris()
X=iris.data #this contains the input features
Y=iris.target #this contains the target labelling like 0,1,2 for the three species mentioned in the given dataset

#Splitting the dataset into training and testing set
X_train, X_test, Y_train, Y_test, = train_test_split(X,Y, test_size=0.2, random_state=42)

#Normalising the values to a standard form
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)


#Now we will be manually hard coding the KNN algorithm
class KNN:
  def __init__(self,k=3):
    self.k=k

  def fit(self,X,Y):      #this function is used to label the training data
    self.X_train=X
    self.Y_train=Y


  def euclidian(self,x1,x2):  #we would be using euclidian distance to calculate the rank
    distance=0.0
    for i in range(len(x1)):
      distance+=(x1[i]-x2[i])**2
    return np.sqrt(distance)


  def predict_single(self,x):     #here we predict the label for a single data
    distances=[]
    for i in range(len(self.X_train)):
      dist=self.euclidian(x, self.X_train[i])    #calculating the distances
      distances.append((dist,self.Y_train[i]))   #stores the distance label pairs
      distances.sort(key=lambda tup:tup[0])      #sorting the distances in ascending order for better view

      k_labels=[]
      for i in range(self.k):
        k_labels.append(distances[i][1])         #taking label of top k closest neighbours
        label_count=Counter(k_labels)
        predicted_label=label_count.most_common(1)[0][0]    #returns the label which occurs most frequently
        return predicted_label

    def predict(self,X):
      predictions=[]
      for x in X:
        label=self.predict_single(x)
        predictions.append(label)
      return np.array(predictions)        #returning results as numpy array


#training and testing KNN
knn=KNN(k=3)                              #creating knn
knn.fit(X_train_scaled,Y_train)           #training on scaled data
Y_pred_knn=knn.predict(X_test_scaled)     #predicting on test data


#logistic regression prediction
logreg=LogisticRegression(max_iter=200)   #creating logisitc regression with max 200 iterations
logreg.fit(X_train_scaled,Y_train)        #training on scaled data
Y_pred_logreg=logreg.predict(X_test_scaled)  #tetsing on scaled test data


#descision tree prediction
dtree=DecisionTreeClassifier()            #creating a decision tree
dtree.fit(X_train_scaled,Y_train)         #training on scaled train data
Y_pred_dtree=dtree.predict(X_test_scaled) #testing on test data


#evaluating the functions
def evaluate_model()


```
